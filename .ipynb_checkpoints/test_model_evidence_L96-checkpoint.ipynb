```python
### GENERATE SIMULATED DATA (LORENZ-96 MODEL)

# parameters
class GD:
    model = 'Lorenz_96'
    class parameters:
        F = 8
        J = 40
    dt_integration = 0.05 # integration time
    dt_states = 1 # number of integration times between consecutive states (for xt and catalog)
    dt_obs = 4 # number of integration times between consecutive observations (for yo)
    var_obs = np.random.permutation(parameters.J)[0:20] # indices of the observed variables
    nb_loop_train = 10**3 # size of the catalog
    nb_loop_test = 10 # size of the true state and noisy observations
    sigma2_catalog = 0   # variance of the model error to generate the catalog   
    sigma2_obs = 2 # variance of the observation error to generate observations
    
# run the data generation
catalog_F8, xt_F8, yo_F8 = AnDA_generate_data(GD)
```

```python
### GENERATE SIMULATED DATA (LORENZ-96 MODEL)

# parameters
class GD:
    model = 'Lorenz_96'
    class parameters:
        F = 11
        J = 40
    dt_integration = 0.05 # integration time
    dt_states = 1 # number of integration times between consecutive states (for xt and catalog)
    dt_obs = 4 # number of integration times between consecutive observations (for yo)
    var_obs = np.random.permutation(parameters.J)[0:20] # indices of the observed variables
    nb_loop_train = 10**3 # size of the catalog
    nb_loop_test = 10 # size of the true state and noisy observations
    sigma2_catalog = 0   # variance of the model error to generate the catalog   
    sigma2_obs = 2 # variance of the observation error to generate observations
    
# run the data generation
catalog_F11, xt_F11, yo_F11 = AnDA_generate_data(GD)
```

```python
from scipy.stats import gaussian_kde

min_lim=min(catalog_F11.analogs[:,20])
max_lim=max(catalog_F11.analogs[:,20])

subplot(1,2,1)
x = catalog_F8.analogs[:,20]
y = catalog_F8.successors[:,20]
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
scatter(x, y, c=z, s=50, edgecolor='')
#axis('equal')
xlim([min_lim,max_lim])
ylim([min_lim,max_lim])
xlabel('$x_{20}(t)$', size=20)
ylabel('$x_{20}(t+$' + str(GD.dt_integration) +'$)$' , size=20)
title('F=8', size=20)
subplot(1,2,2)
x = catalog_F11.analogs[:,20]
y = catalog_F11.successors[:,20]
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
scatter(x, y, c=z, s=50, edgecolor='')
#axis('equal')
xlim([min_lim,max_lim])
ylim([min_lim,max_lim])
xlabel('$x_{20}(t)$', size=20)
ylabel('$x_{20}(t+$' + str(GD.dt_integration) +'$)$' , size=20)
title('F=11', size=20)

# save figure
savefig('/home/ptandeo/Dropbox/Documents/Figures/AnDA_model_evidence/exp_L96_xt_xt_dt.png', bbox_inches='tight', dpi=400)
```
